```python
import os
import shutil

gdrive_project_dir = '/content/gdrive/MyDrive/Projects/flower_category_prediction/'

def prepare_data(target_dir, img_file):
    train_files = open(img_file).read().split('\n')
    print(len(train_files))
    dirs = os.listdir('17FlowerDataset')
    dirs = [dir for dir in dirs if '.' not in dir]
    for dir in dirs:
        file_path = f'17FlowerDataset/{dir}'
        c = 0
        for file in train_files:
            p = f'{file_path}/{file}'
            if os.path.isfile(p):
                target_train = f'{target_dir}/{dir}'
                if not os.path.exists(target_train):
                    os.makedirs(target_train)
                shutil.copy(
                    os.path.join(file_path, file), os.path.join(target_train, file)
                )
                c += 1
        print(c)

prepare_data(f'{gdrive_project_dir}dataset/train', '17FlowerDataset/TrainSet1.txt')
prepare_data(f'{gdrive_project_dir}dataset/test', '17FlowerDataset/TestSet1.txt')
prepare_data(f'{gdrive_project_dir}dataset/validation', '17FlowerDataset/ValSet1.txt')
```
